```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix, ConfusionMatrixDisplay
import joblib
import os
```

```python
df=pd.read_csv('/content/obesity_level.csv')
```

```python
print(df.isnull().sum())
```

```python
df = df.drop_duplicates()
```

```python
for col in df.select_dtypes(include="object").columns:
    df[col] = df[col].str.strip().str.lower()
```

```python
print(df.head())
```

```python
if "bmi" not in df.columns:
    df["bmi"] = df["Weight"] / ((df["Height"]/100)**2)
if "FCVC" in df.columns:
    df["frequent_fastfood"] = (df["FCVC"] > 2).astype(int)


columns_to_print = ["bmi"]
if "frequent_fastfood" in df.columns:
    columns_to_print.append("frequent_fastfood")

print(df[columns_to_print].head())
```

```python
sns.countplot(x="0be1dad", data=df)
plt.title("Distribution of Obesity Levels")
plt.xticks(rotation=45)
plt.show()
```

```python
sns.boxplot(x="0be1dad", y="bmi", data=df)
plt.title("BMI by Obesity Level")
plt.xticks(rotation=45)
plt.show()
```

```python
if "FAF" in df.columns:
    sns.boxplot(x="0be1dad", y="FAF", data=df)
    plt.title("Physical Activity (FAF) vs Obesity Level")
    plt.xticks(rotation=45)
    plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

```python

le = LabelEncoder()
df["target"] = le.fit_transform(df["0be1dad"])
```

```python
X = df.drop(["0be1dad","target"], axis=1)
y = df["target"]
```

```python
X = pd.get_dummies(X, drop_first=True)
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=42)
```

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

print("Train shape:", X_train.shape)
print("Test shape:", X_test.shape)
```

```python
log_reg = LogisticRegression(max_iter=1000, multi_class="multinomial")
log_reg.fit(X_train, y_train)
y_pred_lr = log_reg.predict(X_test)

print("Logistic Regression Report:\n", classification_report(y_test, y_pred_lr))
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt

cm = confusion_matrix(y_test, y_pred_lr)
disp = ConfusionMatrixDisplay(cm, display_labels=le.classes_)
disp.plot(cmap="Blues", xticks_rotation=45)

plt.title("Confusion Matrix - Logistic Regression")
plt.tight_layout()
plt.show()
```

```python
rf = RandomForestClassifier(n_estimators=100, random_state=42, class_weight="balanced")
rf.fit(X_train, y_train)
y_pred_rf = rf.predict(X_test)
print("Random Forest Report:\n", classification_report(y_test, y_pred_rf))
```

```python
cm = confusion_matrix(y_test, y_pred_rf)
plt.figure(figsize=(10, 7))  # enlarge the frame
disp = ConfusionMatrixDisplay(cm, display_labels=le.classes_)
disp.plot(cmap="Greens", xticks_rotation=45, ax=plt.gca())  # rotate labels

plt.title("Confusion Matrix - Random Forest")
plt.tight_layout()  # adjust to prevent cut-off
plt.show()
```

```python
joblib.dump(rf, "obesity_rf_model.joblib")
joblib.dump(scaler, "scaler.joblib")
joblib.dump(le, "label_encoder.joblib")
print("Model and encoders saved!")
```
